```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import scipy
import numpy as np
```

```python
# total
total_label = "B20002_001E"

for year in range(2010, 2021):
    yearStr = str(year)
    df_year = pd.read_excel(yearStr +".xlsx")
    if year == 2010:
        df = df_year.iloc[:,0]
        for i in range(1,4):
            df = pd.concat([df, df_year.iloc[:,i]],axis=1)# columns containing geo info and total median
            df.rename(columns = {total_label: yearStr}, inplace = True)
        df.columns.values[1] = "County"
        df.columns.values[2] = "State"
    else:
        df = pd.concat([df, df_year.iloc[:,3]],axis=1)#
        df.rename(columns = {total_label: yearStr}, inplace = True)
        #df = df.concat(df_year.iloc[:,3],ignore_index=True)

df = df.drop(index=0)
df = df.reset_index(drop=True)
df.head()
#type(df.iloc[0,1])
```

```python
df.to_csv("total_gender_data.csv",index=False)
```
